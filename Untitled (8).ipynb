```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.utils.multiclass import unique_labels
from sklearn.metrics import classification_report,  confusion_matrix , ConfusionMatrixDisplay
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.preprocessing import label_binarize
```

```python
def load_data(): 
    train_loaded = np.load(f'train_data.npz')
    test_loaded = np.load(f'test_data.npz')
    
    train_x, train_y = train_loaded['x'].T, train_loaded['y'] 
    test_x, test_y = test_loaded['x'].T, test_loaded['y'] 

    # Standarize data
    train_x = train_x / 255.
    test_x = test_x / 255.

    # Shuffle data (no effect for full batch method)
    train_indices = np.random.permutation(train_x.shape[1])
    train_x = train_x[:, train_indices]
    train_y = train_y[train_indices]

    test_indices = np.random.permutation(test_x.shape[1])
    test_x = test_x[:, test_indices]
    test_y = test_y[test_indices]

    return train_x, train_y, test_x, test_y

train_x, train_y, test_x, test_y = load_data()
print('train data matrix: ', train_x.shape)
print('test data matrix: ', test_x.shape)

print('number of data samples for each class (train): ', np.unique(train_y, return_counts=True))
print('number of data samples for each class (test): ', np.unique(test_y, return_counts=True))

# Plot some of the images 
def reshape_images(img):
    """
    reshapes the given matrix from [12288, ] to [64, 64, 3]
    """
    return np.reshape(img, (64, 64, 3))

def show_img(X, title: str = ''):
    plt.imshow(X)
    plt.axis('off')
    plt.title(f'{title}')
    plt.show()

index = 0
show_img(reshape_images(train_x[:, index]), str(train_y[index]))
index = 1003
show_img(reshape_images(train_x[:, index]), str(train_y[index]))
```

```python
def sigmoid(z):
  """Sigmoid activation function"""
  return 1 / (1 + np.exp(-z))

def relu(z):
  """ReLU activation function"""
  return np.maximum(0, z)

def cross_entropy_loss(y_true, y_pred):
  """Cross-entropy loss function"""
  return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

class MLP:
  def __init__(self, layers, activations):
    """
    Initializes the MLP network with given layers and activation functions.

    Args:
      layers: List of integers representing the number of neurons in each layer.
      activations: List of activation functions (functions) for each layer.
    """
    self.layers = layers
    self.activations = activations
    self.weights = []
    self.biases = []
    for i in range(1, len(layers)):
      # Xavier initialization for weights
      limit = np.sqrt(6 / (layers[i-1] + layers[i]))
      # Corrected weight initialization for the first hidden layer
      if i == 1:
        self.weights.append(np.random.rand(layers[i], layers[i-1]) * 2 * limit - limit)
      else:
        self.weights.append(np.random.rand(layers[i], layers[i-1]) * 2 * limit - limit)
      self.biases.append(np.zeros((layers[i], 1)))

  def forward(self, X):
    """
    Performs forward pass through the network.

    Args:
      X: Input data (numpy array).

    Returns:
      A dictionary containing activations (A) and weighted sums (Z) for each layer.
    """
    activations = {}
    activations[0] = X
    for i in range(1, len(self.layers)):
      Z = np.dot(self.weights[i-1], activations[i-1]) + self.biases[i-1]
      activations[i] = self.activations[i](Z)
    return activations

  def backward(self, activations, y_true):
    """
    Performs backward pass to calculate gradients for weights and biases.

    Args:
      activations: Dictionary containing activations (A) and weighted sums (Z) for each layer.
      y_true: True labels (numpy array).

    Returns:
      A list of weight gradients and a list of bias gradients.
    """
    gradients_weights = []
    gradients_biases = []
    m = len(y_true)  # number of training samples

    # Output layer
    delta = activations[len(self.layers) - 1] - y_true
    gradients_weights.append(np.dot(delta, activations[len(self.layers) - 2].T) / m)
    gradients_biases.append(np.sum(delta, axis=1, keepdims=True) / m)

    # Hidden layers
    for i in range(len(self.layers) - 2, 0, -1):
      delta = np.dot(self.weights[i].T, delta) * self.activations[i](activations[i]) * (1 - self.activations[i](activations[i]))
      gradients_weights.append(np.dot(delta, activations[i-1].T) / m)
      gradients_biases.append(np.sum(delta, axis=1, keepdims=True) / m)

    return gradients_weights[::-1], gradients_biases[::-1]

  def update(self, learning_rate, gradients_weights, gradients_biases):
    """
    Updates network weights and biases based on gradients and learning rate.

    Args:
      learning_rate: Learning rate for gradient descent.
      gradients_weights: List of weight gradients.
      gradients_biases: List of bias gradients.
    """
    for i in range(len(self.weights)):
      self.weights[i] -= learning_rate * gradients_weights[i]
      self.biases[i] -= learning_rate * gradients_biases[i]

  def predict(self, X):
    """
    Predicts class labels for given data.

    Args:
      X: Input data (numpy array).

    Returns:
      Predicted class labels (numpy array).
    """
    activations = self.forward(X)
    predictions = np.round(activations[len(self.layers) - 1])
    return predictions

  def train(self, X_train, y_train, X_test, y_test, epochs, learning_rate):
        train_loss_history = []
        test_loss_history = []
        train_accuracy_history = []
        test_accuracy_history = []

        for epoch in range(epochs):
            activations_train = self.forward(X_train)
            loss_train = cross_entropy_loss(y_train, activations_train[len(self.layers) - 1])
            train_loss_history.append(loss_train)
            train_accuracy = np.mean(self.predict(X_train) == y_train)
            train_accuracy_history.append(train_accuracy)

            activations_test = self.forward(X_test)
            loss_test = cross_entropy_loss(y_test, activations_test[len(self.layers) - 1])
            test_loss_history.append(loss_test)
            test_accuracy = np.mean(self.predict(X_test) == y_test)
            test_accuracy_history.append(test_accuracy)

            gradients_weights, gradients_biases = self.backward(activations_train, y_train)
            self.update(learning_rate, gradients_weights, gradients_biases)
            print(f"Epoch: {epoch+1}/{epochs}, Train Loss: {loss_train:.4f}, Test Loss: {loss_test:.4f}")

        # Plotting
        epochs_range = range(1, epochs+1)

        plt.figure(figsize=(12, 6))

        plt.subplot(1, 2, 1)
        plt.plot(epochs_range, train_loss_history, label='Train Loss')
        plt.plot(epochs_range, test_loss_history, label='Test Loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()

        plt.subplot(1, 2, 2)
        plt.plot(epochs_range, train_accuracy_history, label='Train Accuracy')
        plt.plot(epochs_range, test_accuracy_history, label='Test Accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()

        plt.tight_layout()
        plt.show()
```

```python
def model_evaluation(clf, x_train, x_test, y_train, y_test, model_name):  
    sns.set(font_scale=1.2)

    print("-"*15,"Performace Model on Testset")
    print("-"*55)
    y_pred=clf.predict(x_test)
    y_pred=y_pred.reshape(400)
    print(classification_report(y_test,y_pred))
    "\n"
    print("-"*15,"Performace Model on Trainset")
    print("-"*55)
    y_pred_train=clf.predict(x_train)
    y_pred_train=y_pred_train.reshape(2000)
    print(classification_report(y_train,y_pred_train))
```

```python
def Confusion_Matrix(model,test_x,test_y):
    y_pred=model.predict(test_x)
    y_pred=y_pred.reshape(400)
    cm = confusion_matrix(test_y, y_pred)

# Create a visualization of the confusion matrix
    plt.figure(figsize=(6, 4))
    class_names = ['Panda (0)', 'Bear (1)']  # Adjust class names if needed
    plt.imshow(cm, cmap='YlOrBr')
    plt.colorbar()

# Add text annotations for each cell
    for i in range(len(class_names)):
        for j in range(len(class_names)):
            plt.text(j, i, cm[i, j], ha='center', va='center', color='black', fontsize=12)  # Add cell text

    plt.xticks(range(len(class_names)), class_names) #, rotation=45
    plt.yticks(range(len(class_names)), class_names)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix for MLP Network')


    plt.show()
```
